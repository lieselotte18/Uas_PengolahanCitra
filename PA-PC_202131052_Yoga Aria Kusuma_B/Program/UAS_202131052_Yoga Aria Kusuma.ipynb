### Import Library

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Deteksi Gambar

```python
img = cv.imread('d.jpg')
```

```python
cv.imshow("contour", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Menampilkan Tepi Pada Gambar

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2RGB)
edges= cv.Canny(img, 30, 150)
#edges digunakan untuk melakukan pendeteksian
#canny digunakan detekesi gambar
```

```python
lines = cv.HoughLinesP(edges, 1, np.pi/180, 30, maxLineGap = 250) #digunakan untuk mendeteksi garis cv.houghlines
img_line = img.copy()                                             
```

```python
for line in lines:
    x1, y1, x2, y2, = line[0]
    cv.line(img_line, (x1,y1), (x2,y2), (0,0,225), 6)
```

### Tampilan Menggunakan Figure & Axis (Fungsi Matplotlib)

```python
gray1 = cv.cvtColor(img_line, cv.COLOR_BGR2RGB)

#Tampilan Menggunakan Index
fig, axs = plt.subplots(1,2, figsize = (10,10))
ax = axs.ravel()

#Tampilan Asli
ax[0].imshow(gray, cmap= 'gray')
ax[0].set_title('Citra Asli')

#Tampilan Edges
ax[1].imshow(gray1, cmap= 'gray')
ax[1].set_title('Setelah Edges')
```

### Membuat Background Gambar Berwarna Putih

```python
img_line = np.ones_like(img) * 255  # Mengubah semua piksel menjadi putih (255)
```

### Membuat Garis Tepi

```python
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv.line(img_line, (x1, y1), (x2, y2), (0, 0, 225), 6)  # Garis tepi berwarna merah (255, 0, 0)
```

```python
gray1 = cv.cvtColor(img_line, cv.COLOR_BGR2RGB)

# Tampilan Menggunakan Index
fig, axs = plt.subplots(1, 2, figsize=(10, 10))
ax = axs.ravel()

# Tampilan Asli
ax[0].imshow(gray, cmap='gray')
ax[0].set_title('Citra Asli')

# Tampilan Edges
ax[1].imshow(gray1, cmap='gray')
ax[1].set_title('Setelah Edges')

plt.show()
```
